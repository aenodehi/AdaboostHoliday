```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y_s = make_classification(n_samples=1000, n_features=2, n_classes=2,
                         n_clusters_per_class=1, n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y_s)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_s, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
svc.coef_
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train, y_train)

y_pred1 = rbf.predict(X_test)

print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial = SVC(kernel='poly')
polynomial.fit(X_train, y_train)

y_pred2 = rbf.predict(X_test)

print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)

y_pred3 = rbf.predict(X_test)

print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

### Hyperparameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}

grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(X_test)

print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
